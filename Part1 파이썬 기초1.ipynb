#### 자료구조

자료구조는 어떤 데이터를 저장할 때, 그 데이터에 특징에 따라 컴퓨터에 효율적으로 정리하기 위한 데이터의 저장 및 표현 방식을 이야기합니다. 어떤 데이터는 순서가 있다거나, 그 데이터의 나타내는 ID 값과 쌍을 이룬다던가 하는 등의 특징이 나타나게 됩니다. 일반적으로 사용되는 정수, 문자열등의 변수 타입보다는 이러한 특징들에 잘 맞는 형태로 데이터를 저장하게 된다면 훨씬 효율적으로 컴퓨터의 메모리를 사용하고, 프로그래머가 코드를 작성하기에도 용이하게 해줍니다.

*자료구조*

1. **배열(Array)**: 배열은 데이터를 순서대로 나열한 것 같아. 예를 들어, 숫자들을 일렬로 나열한 것처럼. [1, 2, 3, 4, 5] 이런 식으로 순서대로 저장해.

2. **리스트(List)**: 리스트는 배열과 비슷하지만, 데이터를 더 자유롭게 추가하고 삭제할 수 있어. 예를 들어, 친구들의 이름을 순서 없이 추가할 수 있는 것처럼. [“철수”, “영희”, “민수”]

3. **스택(Stack)**: 스택은 쌓아 올리는 자료구조야. 새로운 데이터를 위에 쌓고, 맨 위에서부터 꺼내. 마치 접시를 쌓아 올리고, 맨 위에서부터 하나씩 꺼내는 것처럼. 후입선출(LIFO, Last In First Out) 방식이야.

4. **큐(Queue)**: 큐는 줄 서는 자료구조야. 먼저 들어온 데이터가 먼저 나가. 마치 줄을 서서 순서대로 차례를 기다리는 것처럼. 선입선출(FIFO, First In First Out) 방식이야.

5. **트리(Tree)**: 트리는 가지가 많은 나무처럼 생긴 자료구조야. 데이터가 부모와 자식 관계로 연결되어 있어. 예를 들어, 가족나무를 생각하면 쉬워. 할아버지, 할머니가 부모 노드고, 그 아래에 아빠, 엄마가 있고, 또 그 아래에 너와 형제들이 있는 것처럼.

6. **해시 테이블(Hash Table)**: 해시 테이블은 데이터를 키(Key)와 값(Value) 쌍으로 저장해. 예를 들어, 친구의 이름(Key)과 전화번호(Value)를 짝지어 저장하는 것처럼. “철수”: “010-1234-5678”

7. **그래프(Graph)**: 그래프는 여러 점(노드)이 선(간선)으로 연결된 자료구조야. 예를 들어, 친구들 사이의 관계를 나타내는 것처럼. 철수와 영희가 친구고, 영희와 민수가 친구인 것을 선으로 연결해 나타낼 수 있어.

이렇게 다양한 자료구조가 있어! 각 자료구조는 특정 상황에서 더 효율적으로 데이터를 저장하고 사용할 수 있게 도와줘

```python
# 객체라는걸 어떻게 코드로 표현을 하는가?
# 클래스 이름은 Camelcase ^대문자^대문자

class SoccerPlayer(object):
    def __init__(self, name, position, back_number):
        self.name = name
        self.position = position
        self.back_number = back_number

    def __str__(self):
        return "Hello, My name is %s. My back Number is %d" % (self.name, self.back_number)
    
    def __add__(self, other):
        return self.name + other.name
```

```python
abc = SoccerPlayer("son", "FW", 7)
bc = SoccerPlayer("Park", "WF", 13)
# c = SoccerPlayer()
```

```python
# First Class Function 일급함수
def square(x):
    return x * x


def cube(x):
    return x * x * x


def formula(method, argument_list):
    return [method(value) for value in argument_list]

```

```python
formula(cube, [1,2,3,4,5])
```

```python
import sympy as sym
```

```python
from sympy.abc import x

print(sym.diff((x**2 + 2*x + 3),x))
```

```python
import numpy as np

def func(val):
    fun = sym.poly(x**2 + 2*x + 3)
    return fun.subs(x, val), fun


def func_gradient(fun, val):
    _, function = fun(val)
    diff = sym.diff(function, x)
    return diff.subs(x, val), diff


def gradient_descent(fun, init_point, lr_rate=1e-2, epsilon=1e-5):
    cnt = 0
    val = init_point
    diff, _ = func_gradient(fun, init_point)
    while np.abs(diff) > epsilon:
        val = val - lr_rate*diff
        diff, _ = func_gradient(fun, val)
        cnt += 1

    print("함수: {}, 연산횟수: {}, 최소점: ({}, {})".format(fun(val)[1], cnt, val, fun(val)[0]))

gradient_descent(fun=func, init_point=np.random.uniform(-2,2))
```

```python
# Multivariate Gradient Decent
import numpy as np
x,y = sym.symbols('x,y')

def eval_(fun, val):
    val_x, val_y = val
    fun_eval = fun.subs(x, val_x).subs(y, val_y)
    return fun_eval


def func_multi(val):
    fun = sym.poly(x**2 + 2*y**2, x, y)
    return eval_(fun, val), fun


def func_gradient(fun, val):
    x_, y_ = val
    func_val, function = fun(val)
    diff_x = sym.diff(function, x)
    diff_y = sym.diff(function, y)
    grad_vec = np.array([eval_(diff_x, [x_, y_]), eval_(diff_y, [x_, y_])], dtype=float)
    return grad_vec, [diff_x, diff_y]


def gradient_descent(fun, init_point, lr_rate=1e-2, epsilon=1e-5):
    cnt = 0
    val = init_point
    diff, _ = func_gradient(fun, val)
    while np.linalg.norm(diff) > epsilon:
        val = val - lr_rate*diff
        diff, _ = func_gradient(fun, val)
        cnt += 1

    print("함수: {}, 연산횟수: {}, 최소점: ({}, {})".format(fun(val)[1], cnt, val, fun(val)[0]))

pt=[np.random.uniform(-2,2), np.random.uniform(-2,2)]
gradient_descent(fun=func_multi, init_point = pt)
```
